对所有客户的购买产品情况进行统计

```python
import pandas as pd
import os
os.chdir('D:\python data')
```

```python
df = pd.read_csv('order.csv',encoding = 'gbk')
```

```python
df
```

```python
df_1 = df[['客户ID','产品名称']]
```

```python
df_1
```

```python
pt = df_1.groupby(df_1['客户ID'])['产品名称'].value_counts().unstack().reset_index().set_index('客户ID').fillna(0)
#以客户ID作为划分依据，划分结果为每个产品的出现次数(value_counts)，得到每个客户的购买情况
```
